1. Remove HTML content

```python
def remove_names_and_dates(page_html_text):
    parsed_html = BeautifulSoup(page_html_text)
    p_lines = parsed_html.findAll('p')
    
    min_sentense_word_count = 13
    p_index = 0
    
    for p_line in p_lines:
        
        strong_lines = p_line.findAll('strong')
        if not strong_lines:
            continue 
            
        for s in strong_lines:
            if (len(s.text.split(" ")) >= min_sentense_word_count ):
                break
            else:
                p_index += 1
                
                continue
            break
            
        for i in range(0,p_index):
            page_html_text = page_html_text.replace(str(p_lines[i]),"")
            
        return page_html_text
```

```python
data['main_html'].isna().sum()
```

```python
data['main_html_v2'] = data['main_html'].fillna('').apply(lambda x: remove_names_and_dates(x))
```

2. Remove all numbers

```python
import re

pattern = "\[(\d+)\]"


data['main_html_v3'] = data['main_html_v2'].apply(lambda x: re.sub(pattern,'',str(x)))
```

```python
data['main_html_v3'][0]
```

3.Remove list of references

```python
data['main_html_v4'] = data['main_html_v3'].apply(lambda x: BeautifulSoup(x).text)
```

```python
data['main_html_v5'] = data['main_html_v4'].apply(lambda x: re.sub(r'http(\S+.*\s)',"",x))
data['main_html_v6'] = data['main_html_v5'].apply(lambda x: re.sub(r'2022|2023|©2022|©2023|\xa0|\n',"",x))
```
